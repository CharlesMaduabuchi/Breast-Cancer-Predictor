```python
# Missing values
invalid_cols = ['radius_mean', 'texture_mean', 'perimeter_mean', 'area_mean', 'smoothness_mean', 'compactness_mean', 'concavity_mean', 'concave_points_mean',]

df[invalid_cols] = df[invalid_cols].replace(0, np.nan)

df.isnull().sum()
```

```python
sns.countplot(data=df, x='diagnosis')
plt.title("Diagnosis Distribution (M=Malignant, B=Benign)")
plt.show()
```

```python
#Label Encoding
df['diagnosis'] = df['diagnosis'].map({'M': 1, 'B': 0})
#Malignant means cancerous
#Benign means non-cancerous
#In the context of tumors, benign means non-cancerous, while malignant means cancerous. Benign tumors do not invade surrounding tissue or spread to other parts of the body, whereas malignant tumors can grow uncontrollably and spread, potentially harming other tissues and organs. 
```

```python
# Plot the distribution of all numeric features
df.hist(figsize=(15, 10), bins=20, color='skyblue', edgecolor='black')
plt.suptitle('Feature Distributions', fontsize=16)
plt.tight_layout()
plt.show()
```

```python
# Compute correlation matrix
# First, select only numeric columns from the dataframe
numeric_df = df.select_dtypes(include=['float64', 'int64'])

# Use numeric_df instead of df for correlation calculation
plt.figure(figsize=(20, 20))
sns.heatmap(numeric_df.corr(), annot=True, fmt=".2f", cmap='coolwarm', square=True)
plt.title('Correlation Matrix for Breast Cancer')
plt.show()
```

```python
# Create the correlation matrix
corr = df.corr()

# Top correlated features with diagnosis
top_corr = corr['diagnosis'].abs().sort_values(ascending=False)[1:11]
print("Top features correlated with diagnosis:")
print(top_corr)
```

```python
# Select top features
top_features = top_corr.index.tolist()

# Pairplot (slow for many features — pick 4–6)
sns.pairplot(df, vars=top_features, hue='diagnosis')
plt.show()

# Boxplots for comparison
for feature in top_features:
    plt.figure(figsize=(6, 4))
    # Drop any NaN values in the feature column to prevent the error
    valid_data = df.dropna(subset=[feature, 'diagnosis'])
    sns.boxplot(x='diagnosis', y=feature, data=valid_data)
    plt.title(f"{feature} by Diagnosis")
    plt.show()
```

```python
for feature in top_features[:8]:  # Just the top 8 for brevity
    plt.figure(figsize=(6, 4))
    sns.histplot(df[feature], kde=True)
    plt.title(f"Distribution of {feature}")
    plt.show()
```

```python
# First, separate numeric and categorical columns
numeric_cols = df.select_dtypes(include=['number']).columns
categorical_cols = df.select_dtypes(exclude=['number']).columns

# Impute missing values for numeric columns with median
if len(numeric_cols) > 0:
    df[numeric_cols] = df[numeric_cols].fillna(df[numeric_cols].median())

# For categorical columns, impute with mode (most frequent value)
if len(categorical_cols) > 0:
    for col in categorical_cols:
        df[col] = df[col].fillna(df[col].mode()[0])

# Confirm no missing values remain
df.isnull().sum()
```

```python
features = ['radius_mean', 'texture_mean', 'perimeter_mean', 'area_mean', 'smoothness_mean', 'compactness_mean', 'concavity_mean', 'concave_points_mean',]

X = df[features]
y = df.diagnosis
```

```python
# Standardize features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Split the scaled data into 80% training and 20% testing
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.4, random_state=42, stratify=y)
```

```python
model = RandomForestClassifier(n_estimators=500,n_jobs=-1)
model.fit(X_train, y_train)
```

```python
# Predict on test set
y_pred = model.predict(X_test)
```

```python
# Evaluate performance
print("Accuracy:", accuracy_score(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))
```

```python
# Plot confusion matrix
conf_mat = confusion_matrix(y_test, y_pred)

plt.figure(figsize=(6, 4))
sns.heatmap(conf_mat, annot=True, fmt='d', cmap='Blues', xticklabels=[0, 1], yticklabels=[0, 1])
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()
```

```python
while True:
    user_input = input("Enter a test row number to predict (or 'q' to quit): ")
    
    if user_input.lower() == 'q':
        print("Exiting...")
        break
    
    if not user_input.isdigit():
        print("Please enter a valid integer row number or 'q' to quit.")
        continue
    
    row_index = int(user_input)
    
    if row_index < 0 or row_index >= len(X_test):
        print(f"Please enter a number between 0 and {len(X_test) - 1}")
        continue
    
    sample = X_test[row_index].reshape(1, -1)
    predicted_label = model.predict(sample)[0]
    predicted_prob = model.predict_proba(sample)[0][1]
    actual_label = y_test.iloc[row_index]
    
    result = "Cancerous" if predicted_label == 1 else "Non-Cancerous"
    
    print(f"Row {row_index} prediction: {result} (probability of Breast Cancer = {predicted_prob:.4f})")
    print(f"Actual label: {'Cancerous' if actual_label == 1 else 'Non-Cancerous'}\n")
```
